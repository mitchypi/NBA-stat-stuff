```python
import pandas as pd
import numpy as np
import csv
import matplotlib.pyplot as plt
```

```python
plt.close("all")
```

```python

def poggers():
    csv_file = open("Desktop/stats.csv",'r')
    players = []
    per = []
    ts = []
    tpr = []
    ftr = []
    usg = []
    list1 = []
    csv_file.readline()

    for a, b, c, d, e, f in csv.reader(csv_file, delimiter=','):
        players.append(a)
        per.append(b)
        ts.append(c)
        tpr.append(d)
        ftr.append(e)
        usg.append(f)
    list1.append(players)
    list1.append(per)
    list1.append(ts)
    list1.append(tpr)
    list1.append(ftr)
    list1.append(usg)
    return list1

```

```python
listy = poggers()
#0 = players, 1 = per, 2 = ts, 3 = tpr, 4 = ftr, 5 = usg,
```
